```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris


from sklearn.model_selection import train_test_split # 나나 주기 위해 필요 !
# 결정트리  모델
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import tree

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
```

```python
x =  iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

```

```python
print(feature_names)
```

```python
print(target_names)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(x_train, y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
accurany =  accuracy_score(y_test, y_pred)
coef_metrics = confusion_matrix(y_test, y_pred)
```

```python
class_report  =  classification_report(y_test, y_pred,target_names=target_names)
```

```python
print("Ac" , accurany)
print(coef_metrics)
print(class_report)
```

```python
plt.figure(figsize=(10, 10))
tree.plot_tree(clf, feature_names=feature_names, class_names=target_names, filled=True  )
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x)
labels = kmeans.labels_
```

```python
plt.scatter(x[:, 0], x[:, 1], c=labels,cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Means Clustering')
plt.show()
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression #선형 모델
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import matplotlib.pyplot as plt
```

```python
california = fetch_california_housing()
california_df = pd.DataFrame(california.data, columns=california.feature_names)
```

```python
x=california.data
y= california.target
```

```python
target_names  =  california.target_names
target_names
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
models = { "Random Forest": RandomForestRegressor(n_estimators=100,  random_state=42),"Linear Regression": LinearRegression()}
```

```python
results = {}
```

```python
for name , model in models.items():
  model.fit(x_train,y_train)
  y_pred = model.predict(x_test)
  mse = mean_squared_error(y_test, y_pred)
  r2 = r2_score(y_test, y_pred)
  results[name] = {"MSE": mse, "R2": r2}
  print(f"mean_squared_error : {mse}")
  print(f"r2_score : {r2}")
  print()
  plt.figure(figsize=(10, 6))
  plt.scatter(y_test, y_pred , alpha=0.5, edgecolors="k")
  plt.plot(min(y_test),max(y_test),min(y_test),max(y_test), color= "red", linestyle="-")
  plt.xlabel("actual house val")
  plt.ylabel("predicted house val")
  plt.title(f"{name} Regression Results")
  plt.grid(True)
  plt.show()
print(results)
```

```python
if "Random Forest" in models:
  rf_model = models["Random Forest"]
  feature_importances = rf_model.feature_importances_
  indices = np.argsort(feature_importances)[::-1]
  plt.figure(figsize=(10, 6))
  plt.title("Feature Importance")
  plt.bar(range(x.shape[1]), feature_importances[indices], align="center")
  plt.xticks(range(x.shape[1]), [california.feature_names[i] for i in indices], rotation=90)
  plt.xlim([-1, x.shape[1]])
  plt.xlabel("Features")
  plt.ylabel("Importance")
  plt.grid(True)
  plt.show()
```
