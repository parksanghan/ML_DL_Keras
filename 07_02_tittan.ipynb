```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing # 주택 가격
from sklearn.model_selection import train_test_split # 트레이닝 데이터 분할
from sklearn.linear_model import LinearRegression # 선형 회귀
from sklearn.metrics import mean_squared_error , r2_score
```

mean_squared_error: **평균 제곱 오차(MSE)**를 계산하는 함수입니다. 모델의 예측 값과 실제 값 간의 차이를 제곱한 후 평균을 낸 값으로, 값이 작을수록 모델이 데이터를 잘 예측한 것입니다.
r2_score: **결정 계수(R²)**를 계산하는 함수입니다. R²는 모델이 데이터를 얼마나 잘 설명하는지를 나타내는 지표로, 1에 가까울수록 모델이 데이터를 잘 설명합니다. R²가 0이면 모델이 전혀 설명하지 못하는 것입니다.

LogisticRegression:

로지스틱 회귀는 이진 분류(예: 스팸/비스팸) 문제를 해결하는 데 자주 사용됩니다. 선형 회귀와 유사하지만, 출력값이 0과 1 사이의 확률로 변환되어 클래스(0 또는 1)로 분류됩니다.
SVC (Support Vector Classifier):

**서포트 벡터 머신(SVM)**은 데이터를 분리하는 최적의 경계를 찾는 알고리즘입니다. 선형 또는 비선형 데이터를 분류하는 데 사용되며, 클래스 사이의 여백을 최대화하는 것이 목표입니다.
LinearSVC:

SVM의 변형으로 선형 분류 문제에 사용됩니다. 데이터가 선형적으로 구분 가능할 때 성능이 좋습니다.
RandomForestClassifier:

랜덤 포레스트는 여러 개의 결정 트리(Decision Trees)를 결합하여 분류를 수행하는 앙상블 학습 방법입니다. 개별 결정 트리보다 과적합이 덜하고, 더 높은 정확도를 기대할 수 있습니다.
KNeighborsClassifier:

K-최근접 이웃(K-Nearest Neighbors, KNN) 알고리즘은 새로운 데이터 포인트를 그 주변에 있는 K개의 이웃들과 비교하여 분류합니다. 거리가 가까운 이웃들에 따라 클래스를 결정합니다.
GaussianNB:

**가우시안 나이브 베이즈(Gaussian Naive Bayes)**는 조건부 독립성을 가정하는 단순한 확률 기반 분류기입니다. 데이터를 가우시안(정규 분포)로 모델링하여 확률적으로 분류합니다.
Perceptron:

퍼셉트론은 가장 기본적인 형태의 인공 신경망입니다. 선형 분류 문제를 해결하는 데 사용되며, 한 개의 뉴런을 사용하여 데이터를 학습합니다.
SGDClassifier:

**확률적 경사 하강법(Stochastic Gradient Descent)**을 사용하는 선형 분류 모델입니다. 큰 데이터셋을 빠르게 학습할 수 있도록 최적화된 알고리즘입니다.
DecisionTreeClassifier:

**결정 트리(Decision Tree)**는 데이터의 특징을 기준으로 분할하여 분류하는 트리 기반 모델입니다. 트리의 각 노드는 특정 특징에 대해 조건을 검사하고, 그 결과에 따라 데이터를 분류합니다.

```python
data["Sex"] =  data["Sex"].map({"male": 0 , "female":1})
data["Embarked"]=  data["Embarked"].map({"C":9,"Q":1,"S":2})


data["Age"] =  data["Age"].fillna(data["Age"].median())
data["Embarked"] =  data["Embarked"].fillna(data["Embarked"].mode()[0])
# data["Embarked"] =  data["Embarked"].isna().sum()
x = data[["Pclass","Sex","Age", "Fare","Embarked"]]
y = data["Survived"]
```

```python
# data["Embarked"] =  data["Embarked"].fillna(data["Embarked"].mode()[0])
data["Embarked"].isna().sum()
```

```python
x_train ,  x_test ,  y_train,  y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(len(x_train),len(x_test))
print(len(y_train),len(y_test))
```

```python
model =  LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)
print(model.coef_)
print(model.intercept_)
```

```python
print(y_test)
```

```python
r2_score(y_test,y_pred)
```

```python
error = mean_squared_error(y_test,y_pred)
print(error)
```

```python
from sklearn.metrics import accuracy_score
accurancy  = accuracy_score(y_test,y_pred)
print(accurancy)
```
